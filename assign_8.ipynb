```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

import re
```

```python
data_train = pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/train.csv")
data_test = pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/test.csv")
```

```python
#QUESTION1:
data = pd.concat([data_train,data_test])
data.head(5)
```

```python
#QUESTION2:
data['Survived'] = data['Survived'].dropna()
data['Survived'].value_counts()
```

```python
#QUESTION3:
data['Has_Cabin'] = data['Cabin'].isnull()
data.head(5)
```

```python
#ques.4.
data['Title'] = data.Name.apply(lambda x: re.search(' ([A-Z][a-z]+)\.', x).group(1))
data['Title'] = data['Title'].replace({'Mlle':'Miss', 'Mme':'Mrs', 'Ms':'Miss'})
data['Title'] = data['Title'].replace(['Don', 'Dona', 'Rev', 'Dr',
                                            'Major', 'Lady', 'Sir', 'Col', 'Capt', 'Countess', 'Jonkheer'],'Special')
data.head()
```

```python
#QUESTION5:
data['Age'] = data.Age.fillna(data.Age.median())
data['Fare'] = data.Fare.fillna(data.Fare.median())
data['Embarked'] = data['Embarked'].fillna(method = 'ffill')
data['Cabin'] = data['Cabin'].fillna(method='ffill')
data['Cabin'] = data['Cabin'].fillna(value='C85',limit=1)
data.head(5)
```

```python
#ques.6.
data['Categorical_age']= pd.qcut(data.Age,4, labels=False)
data['Categorical_fare']= pd.qcut(data.Fare, 4, labels=False)
data.head()
```

```python
#ques.7.
data.drop(columns=['Cabin', 'Name', 'PassengerId', 'Ticket', 'Age', 'Fare', 'SibSp', 'Parch'], axis=1, inplace=True)
data.head()
```

```python
#ques.8.
df=pd.get_dummies(data)
df.head()
```
